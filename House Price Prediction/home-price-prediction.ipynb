```python
# Load the dataset
data = pd.read_csv("/home/gautam/Documents/Datasets/House Price Prediction/pred_dataset.csv")
data
```

```python
train_features = ['LotArea', 'YearBuilt', 'OverallQual', 'OverallCond', 'YearRemodAdd', 'BedroomAbvGr', 'Neighborhood', 'PoolArea', 'OpenPorchSF', 'WoodDeckSF', 'GarageArea', 'GarageYrBlt', 'GrLivArea', 'FullBath']

# data[train_features]
# data[train_features].dropna(subset = ['LotArea', 'YearBuilt', 'OverallQual', 'OverallCond', 'YearRemodAdd', 'BedroomAbvGr', 'Neighborhood', 'PoolArea', 'OpenPorchSF', 'WoodDeckSF', 'GarageArea', 'GarageYrBlt', 'GrLivArea', 'FullBath'], inplace=True)
a = data.dropna(subset=['LotArea', 'YearBuilt', 'OverallQual', 'OverallCond', 'YearRemodAdd', 'BedroomAbvGr', 'Neighborhood', 'PoolArea', 'OpenPorchSF', 'WoodDeckSF', 'GarageArea', 'GarageYrBlt', 'GrLivArea', 'FullBath'])
a
```

```python
X = pd.get_dummies(a)
X
```

```python
y = X.SalePrice
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train a random forest classifier on the training data
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model's performance
acc = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(acc * 100))
```
